```python
# Importing Libraries
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
from selenium.webdriver.common.by import By
import re
import pandas as pd

# Initialize the browser
driver = webdriver.Chrome()
driver.maximize_window()

# Open Google Maps
driver.get("https://digit.in/")

time.sleep(2)
```

```python
driver.find_element(By.XPATH,"//div[@class='listing_container']//ul//li[9]").click()
time.sleep(3)
```

```python
#Create empty lists for the features
Laptop_Name = []
Processor = []
Display = []
Operating_sys = []
Memory = []
Graph_proc = []
Body = []
Price = []

```

```python
# Scraping the laptop names
laptop_name= driver.find_elements(By.XPATH,'//div[@class="left_side"]//h3')
for name in laptop_name:
    Laptop_Name.append(name.text)
    
#Scraping the operating system data
try:
    op_sys = driver.find_elements(By.XPATH,"//div[@class='product-detail']/div/ul/li[1]/div/div")
    for o in op_sys:
        Operating_sys.append(o.text)
except NoSuchElementException:
    pass

#Scraping data for the display of the laptop
try:
    display= driver.find_elements(By.XPATH,"//div[@class='product-detail']/div/ul/li[2]/div/div")
    for d in display:
        Display.append(d.text)
except NoSuchElementException:
    pass


#Scraping data for the Processor of the laptop
try:
    processor = driver.find_elements(By.XPATH,"//div[@class='Spcs-details'][1]/table/tbody/tr[5]/td[3]")
    for p in processor:
        Processor.append(p.text)
except NoSuchElementException:
    pass
#Scraping data for memory of the laptop
try:
    memory = driver.find_elements(By.XPATH,"//div[@class='Spcs-details'][1]/table/tbody/tr[5]/td[3]")
    for m in memory:
        Memory.append(m.text)
except NoSuchElementException:
    pass

#Scraping the dimensions and weight of the laptop
try:
    dim_weight = driver.find_elements(By.XPATH,"//div[@class='Spcs-details'][1]/table/tbody/tr[7]/td[3]")
    for dim in dim_weight:
        Body.append(dim.text)
except NoSuchElementException:
    pass


#Scraping the Graph processor of the laptop
try:
    graph = driver.find_elements(By.XPATH,"//div[@class='Spcs-details'][1]/table/tbody/tr[6]/td[3]")
    for g in graph:
        Graph_proc.append(g.text)
except NoSuchElementException:
    pass


#Scraping the prices of the laptop
try:
    price = driver.find_elements(By.XPATH,"//td[@class='smprice']")
    for p in price:
        Price.append(p.text.replace('₹','Rs '))
except NoSuchElementException:
    pass
```

```python
print(len(Laptop_Name),len(Processor),len(Display),len(Operating_sys),len(Memory),len(Graph_proc),len(Body),len(Price))
```

```python
#Create Dataframe and display result
Gaming_Laptop=pd.DataFrame({})
Gaming_Laptop['Laptop Name'] = Laptop_Name
Gaming_Laptop['Processor'] = Processor
Gaming_Laptop['Display'] = Display
Gaming_Laptop['Operating System'] = Operating_sys
Gaming_Laptop['Memory'] = Memory
Gaming_Laptop['Graphics Processor'] = Graph_proc
Gaming_Laptop['Body(Dimension & Weight)'] = Body
Gaming_Laptop['Price'] = Price


Gaming_Laptop
```

```python
#create empty lists
name = []
distance = []
ratings = []
total_reviews = []
overall_reviews = []
privates_price = []
dorms_price = []
facilities = []
description = []

for hostel in hostels:
    name.append(hostel.find_elements(By.XPATH, '//div[@class="property-name"]'.text) 
    distance.append(hostel.find_elements(By.XPATH, '//span[@class="distance-description"]'.text)
    ratings.append(hostel.find_elements(By.XPATH, '//span[@class="number"]'.text)
    total_reviews.append(hostel.find_elements(By.XPATH, '//div[@class="review"]'.text)
    overall_reviews = hostel.find_elements(By.XPATH, '//span[@class="keyword"]'.text
    privates_price.append(hostel.find_elements(By.XPATH, '//div[@class="accommodation-label"]'.text)
    dorms_price.append(hostel.find_elements(By.XPATH, '//div[@class="accommodation-label"]'.text)
    facilities.append(", ".join([item.text for item in hostel.find_elements(By.XPATH, ".//div[@class='facilities-label']")]))
    description.append(hostel.find_element(By.XPATH, '/html/body/div[3]/div/div/div[2]/section/div[9]/div/div[2]/div[4]/div/div[2]').text)

```

```python
print(len(name),len(distance),len(ratings),len(total_reviews),len(overall_reviews),len(privates_price),len(dorms_price),len(facilities),len(description))
```
